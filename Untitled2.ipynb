```python
import pandas as pd

# Load dataset (use latin-1 encoding because utf-8 gives error for this file)
df = pd.read_csv("spam.csv", encoding="latin-1")

# Keep only the first 2 columns: v1 (label), v2 (text)
df = df[['v1', 'v2']]

# Rename columns for clarity
df.columns = ['Label', 'Text']

# Check first 5 rows
print(df.head())
```

```python
print("Total messages:", len(df))
print(df['Label'].value_counts())
```

```python
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r"http\S+|www\S+", "url", text)   # replace links
    text = re.sub(r"\d+", "number", text)          # replace numbers
    text = re.sub(r"[^\w\s]", "", text)            # remove punctuation
    words = text.split()
    words = [word for word in words if word not in stopwords.words("english")]
    return " ".join(words)

df['Clean_Text'] = df['Text'].apply(preprocess_text)

print(df.head())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=5000)
X = vectorizer.fit_transform(df['Clean_Text']).toarray()

y = df['Label'].map({'ham':0, 'spam':1}).values
```

```python
from imblearn.over_sampling import RandomOverSampler

X_resampled, y_resampled = RandomOverSampler(random_state=42).fit_resample(X, y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.2, random_state=42
)

```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report

models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "SVM": SVC(kernel='linear', probability=True),
    "Random Forest": RandomForestClassifier(),
    "AdaBoost": AdaBoostClassifier(),
    "XGBoost": XGBClassifier(eval_metric='logloss')
}
```

```python
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f"\n{name}")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

for name, model in models.items():
    y_pred = model.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Ham', 'Spam'])
    disp.plot(cmap=plt.cm.Blues)
    plt.title(f"Confusion Matrix - {name}")
    plt.show()
```

```python
from sklearn.metrics import roc_curve, auc
import numpy as np

plt.figure(figsize=(8,6))

for name, model in models.items():
    if hasattr(model, "predict_proba"):
        y_score = model.predict_proba(X_test)[:,1]
    else:
        y_score = model.decision_function(X_test)

    fpr, tpr, _ = roc_curve(y_test, y_score)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f"{name} (AUC = {roc_auc:.2f})")

plt.plot([0,1], [0,1], 'k--')  # diagonal line
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison")
plt.legend()
plt.show()
```

```python
import re
import random
import pandas as pd
from sklearn.metrics import accuracy_score

# -------------- 1) Adversarial generator ----------------
def generate_variant_text(msg):
    # common obfuscations attackers use
    subs = [
        (r"\bclick\b", "cl!ck"),
        (r"\bverify\b", "v3rify"),
        (r"\baccount\b", "acc0unt"),
        (r"\bupdate\b", "upd4te"),
        (r"\bpassword\b", "p@ssw0rd"),
        (r"\bbank\b", "b@nk"),
        (r"\bconfirm\b", "c0nf1rm"),
        (r"\burgent\b", "urg3nt"),
        (r"\bfree\b", "fr33"),
        (r"\bwin\b", "w!n"),
        (r"\bclaim\b", "cl@1m"),
        (r"\botp\b", "0tp")
    ]
    text = msg
    # apply several random substitutions to emulate real obfuscation
    for pattern, repl in subs:
        if random.random() < 0.6:  # 60% chance to apply each obfuscation
            text = re.sub(pattern, repl, text, flags=re.IGNORECASE)
    # also sometimes add a benign-looking sentence to dilute signals
    if random.random() < 0.3:
        text = text + " Please confirm receipt."
    return text

# -------------- 2) Prepare test sample (real spam messages) --------------
# Take up to N spam examples from your dataset
N = 500
spam_samples = df[df['Label'].str.lower() == 'spam']['Text'].sample(n=min(N, df[df['Label']=='spam'].shape[0]), random_state=42).tolist()

# If dataset is small or you want to evaluate on all spam, adjust N accordingly

# -------------- 3) Create original and adversarial lists --------------
orig_texts = spam_samples
adv_texts = [generate_variant_text(t) for t in orig_texts]

# -------------- 4) Preprocess (use your preprocess_text function) --------------
orig_clean = [preprocess_text(t) for t in orig_texts]
adv_clean  = [preprocess_text(t) for t in adv_texts]

# -------------- 5) Vectorize --------------
X_orig = vectorizer.transform(orig_clean).toarray()
X_adv  = vectorizer.transform(adv_clean).toarray()

# -------------- 6) Evaluate each model --------------
results = []
examples = {}  # store some fooled examples per model

for name, model in models.items():
    # predictions on original spam (we expect the model to predict 1)
    pred_orig = model.predict(X_orig)
    pred_adv  = model.predict(X_adv)

    # detection rates (fraction of messages predicted as spam)
    detect_orig = pred_orig.sum() / len(pred_orig)
    detect_adv  = pred_adv.sum() / len(pred_adv)
    drop = detect_orig - detect_adv

    results.append({
        'model': name,
        'orig_detection_rate': detect_orig,
        'adv_detection_rate': detect_adv,
        'drop': drop
    })

    # find examples where original was detected as spam but adversarial was NOT detected
    fooled_idx = [i for i,(o,a) in enumerate(zip(pred_orig, pred_adv)) if o==1 and a==0]
    sample_fooled = []
    for i in fooled_idx[:5]:  # up to 5 examples
        sample_fooled.append({
            'original': orig_texts[i],
            'adversarial': adv_texts[i]
        })
    examples[name] = sample_fooled

# -------------- 7) Show results --------------
results_df = pd.DataFrame(results).sort_values('drop', ascending=False)
print("Adversarial robustness results (higher drop = more vulnerable):\n")
print(results_df.to_string(index=False, float_format='%.3f'))

# -------------- 8) Show a few fooled examples for the most vulnerable model --------------
most_vulnerable = results_df.iloc[0]['model']
print(f"\n\nExamples that fooled the model: {most_vulnerable}\n")
for ex in examples[most_vulnerable]:
    print("ORIGINAL : ", ex['original'])
    print("ADVERSARIAL: ", ex['adversarial'])
    print("---")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, roc_curve, auc

for name, model in models.items():
    # Predictions
    y_pred = model.predict(X_test)

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(5,4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f"Confusion Matrix - {name}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

    # ROC Curve
    if hasattr(model, "predict_proba"):  # only for models with probability output
        y_prob = model.predict_proba(X_test)[:,1]
        fpr, tpr, _ = roc_curve(y_test, y_prob)
        roc_auc = auc(fpr, tpr)

        plt.figure(figsize=(5,4))
        plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'AUC = {roc_auc:.2f}')
        plt.plot([0,1], [0,1], color='navy', lw=2, linestyle='--')
        plt.title(f"ROC Curve - {name}")
        plt.xlabel("False Positive Rate")
        plt.ylabel("True Positive Rate")
        plt.legend(loc="lower right")
        plt.show()
```
